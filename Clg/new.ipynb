```python
import numpy as np
import pandas as pd
#create arrray
data=np.array([[3,7],[-4,-6],[1,-1],[7,8],[-4,-1],[-3,-7]])
dataframe=pd.DataFrame(data,columns=['x1','x2'])
```

```python
dataframe
```

```python
dataframe.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
datascaled=scaler.fit_transform(data)
dataframe=pd.DataFrame(datascaled)
dataframe.describe()
```

```python
dataframe=pd.DataFrame(datascaled,columns=['x1','x2'])
dataframe.describe()
```

```python
cov_matrix=dataframe.T @ dataframe / 5
cov_matrix
```

```python
# eigun val eigon vec
eigenvalue , eigenvector = np.linalg.eig(cov_matrix)
eigenvalue
```

```python
pc1=dataframe@ eigenvector[:,0]
pc2=dataframe@ eigenvector[:,1]
pc1
```

```python
data=pd.read_csv('diabetes.csv')
data.head()

data.describe
```

```python
data=data.drop(columns=['Outcome'], axis=1)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
datascaled=scaler.fit_transform(data)
datascaled=pd.DataFrame(datascaled)
datascaled.describe()
```

```python
cov_matrix=datascaled.T @ datascaled / 767
cov_matrix
```

```python
eigenvalue , eigenvector = np.linalg.eig(cov_matrix)
eigenvalue
```

```python
pc1=datascaled@ eigenvector[:,0]
pc2=datascaled@ eigenvector[:,1]
pc3=datascaled@ eigenvector[:,7]
pc1,pc2,pc3
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit_transform(datascaled)
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
plt.plot(pca.explained_variance_ratio_)
```
