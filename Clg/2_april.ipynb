```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
dataframe=load_wine()
data=load_wine()

wine_df=pd.DataFrame(data.data,columns=data.feature_names)

wine_df['target']=data.target

x_train,x_test,y_train,y_test=train_test_split(data.data,data.target,test_size=0.2,random_state=42)

clf=DecisionTreeClassifier(max_depth=4,min_samples_split=50,random_state=42)

clf.fit(x_train,y_train)


y_pred=clf.predict(x_test)

accuracy=accuracy_score(y_test,y_pred)

print(accuracy)
#ploting decision tree
plt.figure(figsize=(14,10))
tree.plot_tree(clf,filled=True,feature_names=data.feature_names,class_names=data.target_names,fontsize=10)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

data=load_wine()

wine_df=pd.DataFrame(data.data,columns=data.feature_names)
# wine_df['target']=data.target
print(wine_df.head())

x_train,x_test,y_train,y_test=train_test_split(data.data,data.target,test_size=0.2,random_state=42)

clf=RandomForestClassifier()
```

```python
param_grid={
    'n_estimators':[50,100,150],
    'max_depth':[2,5,10],
    'min_samples_split':[2,5,10],
    'criterion':['gini','entropy']
}
```

```python
from sklearn.model_selection import GridSearchCV

# Perform the grid search CV
grid_search=GridSearchCV(estimator=clf,param_grid=param_grid,cv=5,scoring='accuracy',n_jobs=-1)
grid_search.fit(x_train,y_train)
```

```python
print("Best parameters found: ", grid_search.best_params_)
```

```python
# Evaluate the model on the test set
best_model=grid_search.best_estimator_
y_pred=best_model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("Test set accuracy: ", accuracy)
print(classification_report(y_test,y_pred))
```
