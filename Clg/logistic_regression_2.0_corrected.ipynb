```python
# need to mask the outliners 
cols = ["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]
for col in cols:
    Q1 = x[col].quantile(0.25)
    Q3 = x[col].quantile(0.75)
    IQR = Q3-Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3+ 1.5*IQR
    mask = (x[col]>= lower_bound) & (x[col] <= upper_bound)
```

```python
# filter the dataset after removal of outliners data
x_outliners_detection = x[mask]
y_outliners_detection = y[mask]
print(type (x_outliners_detection))
print(type (y_outliners_detection))
```

KNN
1. find the optimal value of K.
2. Identification of K neighbours ->
   (i) -> Euclearian sqrt((x2-x1)^2 + (y2-y1)^2)
   (ii) -> k = total number of coordinates (abs(p2 - p1)) {High       Dimensinal data}
   (iii) -> Hemmin Distance -> (binary data or String data) (Same value   = 0) and (different 1) and then addition of all 1
   Find N closest Point

3. Find the values of K neighbours. (Normally we take odd value)
4. unknown (i) -> Regression (mean) and (ii) -> Classification (probability).

5. Find the value.
6. K value is Low (Local Pattern (Overfitting)).
7. K value is High (Generalize (Underfitting)).

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report
from sklearn.neighbors import KNeighborsClassifier

# Create lists to store metrics for different k values
accuracies = []
precisions = []
recalls = []

# Loop through different k values (e.g., from 1 to 20)
for k in range(1, 21):
    # Create KNN classifier with different values of k
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # Fit the classifier on the resampled training data
    knn.fit(x_train_resamples, y_train_resampled)
    
    # Predict on the test set
    y_pred_knn = knn.predict(x_test_knn)
    
    # Calculate and store metrics
    accuracies.append(accuracy_score(y_test, y_pred_knn))
    precisions.append(precision_score(y_test, y_pred_knn))
    recalls.append(recall_score(y_test, y_pred_knn))
    
    # # Print classification report for each k
    # print(f"Classification Report for k={k}:")
    # print(classification_report(y_test, y_pred_knn, target_names=['Non-Diabetic', 'Diabetic']))

# Plot accuracy, precision, and recall on a single graph
plt.figure(figsize=(10, 6))

# Plot each metric on the same graph
plt.plot(range(1, 21), accuracies, marker='o', color='b', label='Accuracy')
plt.plot(range(1, 21), precisions, marker='s', color='g', label='Precision')
plt.plot(range(1, 21), recalls, marker='^', color='r', label='Recall')

# Add titles and labels
plt.title('Accuracy, Precision, and Recall vs. Number of Neighbors (k)')
plt.xlabel('k (Number of Neighbors)')
plt.ylabel('Metric Value')
plt.legend()  # Add a legend to distinguish between metrics
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.neighbors import KNeighborsRegressor

# Create lists to store metrics for different k values
mse_values = []
mae_values = []
r2_values = []

# Loop through different k values (e.g., from 1 to 20)
for k in range(1, 21):
    # Create KNN regressor with KD-Tree for neighbor search
    knn = KNeighborsRegressor(n_neighbors=k, algorithm='kd_tree')
    
    # Fit the regressor on the training data
    knn.fit(x_train, y_train)

    # Predict the values using KNN on normalized data
    y_pred_knn = knn.predict(x_test_knn)
    
    # Calculate and store metrics
    mse_values.append(mean_squared_error(y_test, y_pred_knn))
    mae_values.append(mean_absolute_error(y_test, y_pred_knn))
    r2_values.append(r2_score(y_test, y_pred_knn))

# Plot MSE, MAE, and R² on a single graph
plt.figure(figsize=(10, 6))

# Plot each metric on the same graph
plt.plot(range(1, 21), mse_values, marker='o', color='b', label='MSE (Mean Squared Error)')
plt.plot(range(1, 21), mae_values, marker='s', color='g', label='MAE (Mean Absolute Error)')
plt.plot(range(1, 21), r2_values, marker='^', color='r', label='R² Score')

# Add titles and labels
plt.title('MSE, MAE, and R² Score vs. Number of Neighbors (k) for Regression')
plt.xlabel('k (Number of Neighbors)')
plt.ylabel('Metric Value')
plt.legend()  
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()

print(mse_values)
print(mae_values)
print(r2_values)
```
