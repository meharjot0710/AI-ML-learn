```python
dataframe['Pregnancies']=dataframe['Pregnancies'].replace(0,dataframe['Pregnancies'].median())
dataframe['Glucose']=dataframe['Glucose'].replace(0,dataframe['Glucose'].mean())
dataframe['BloodPressure']=dataframe['BloodPressure'].replace(0,dataframe['BloodPressure'].mean())
dataframe['SkinThickness']=dataframe['SkinThickness'].replace(0,dataframe['SkinThickness'].median())
dataframe['Insulin']=dataframe['Insulin'].replace(0,dataframe['Insulin'].median())
dataframe['BMI']=dataframe['BMI'].replace(0,dataframe['BMI'].mean())
dataframe['DiabetesPedigreeFunction']=dataframe['DiabetesPedigreeFunction'].replace(0,dataframe['DiabetesPedigreeFunction'].median())
dataframe['Age']=dataframe['Age'].replace(0,dataframe['Age'].median())
```

```python
# Take the values for x and y based on correlation matrix
x=dataframe.drop(columns='Outcome', axis=1)
y=dataframe['Outcome']
print(x.columns)
```

```python
fig, ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=x, ax=ax)
plt.savefig('outliners.jpg')
```

The insulin has to be treated as there are many outliers. The difference between 2nd and 3rd quartile is very large which is not good
The ideal one is glucose

```python
x.describe()
```

```python
#need to mask the outliners
cols=['Pregnancies','Glucose','Insulin','BloodPressure','SkinThickness','BMI','DiabetesPedigreeFunction','Age']
for col in cols:
    Q1=x[col].quantile(0.25)
    Q3=x[col].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    mask= (x[col]>=lower_bound)&(x[col]<=upper_bound)

print(x.shape)
```

```python
#filter the dataset after removal
x_outliner_detection=x[mask]
y_outliner_detection=y[mask]
print(x_outliner_detection.shape)
print(y_outliner_detection.shape)
```

```python
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=x_outliner_detection,ax=ax)
plt.savefig('outlinears.jpg')
plt.show()
```

```python
q=x_scaled['Insulin'].quantile(0.95)
mask=x_scaled['Insulin']<q
data_new=x_scaled[mask]
print(data_new.shape)
y_outliner_detection=y_outliner_detection[mask]
print(y_outliner_detection.shape)
```

```python
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=data_new,ax=ax)
plt.savefig('outlinears.jpg')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler

x_train, x_test, y_train, y_test=train_test_split(data_new,y_outliner_detection,test_size=33,random_state=42)

under_sample=RandomUnderSampler(random_state=42)
x_train_resampled_us, y_train_resampled_us=under_sample.fit_resample(x_train, y_train)
```

```python
print(y_train_resampled_us.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# oversampling

# Train Logistic Regression
log_reg = LogisticRegression(random_state=42)
log_reg.fit(x_train_resampled_us, y_train_resampled_us)

# Make predictions on test data
y_pred = log_reg.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy ",accuracy)
```
