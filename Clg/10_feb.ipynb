```python
# import all the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
data={
    "advertisement" : [1.5, 3.0, 4.5, 6.0, 7.5, 9.0,10.5,12.0,13.5,15],
    "sales" : [20, 60, 90, 150,170, 200,240,300, 380,450]
}

df=pd.DataFrame(data)
```

```python
x=df[["advertisement"]].values
y=df["sales"].values

print(x,y)
```

```python
# polyomial features
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
```

```python
# Run the model
model_1=LinearRegression()
model_1.fit(x_poly, y)
# Prediction of model
y_pred = model_1.predict(x_poly)
```

```python
# Linear regression line
model_2 = LinearRegression()
model_2.fit(x,y)

# prediction of Linear regression model
y_pred_2=model_2.predict(x)
```

```python
plt.scatter(x,y,color='blue',label='actual data')
plt.plot(x, y_pred, color='red',label='polynomial regression line')
plt.plot(x, y_pred_2, color='green',label='Linear regression line')
plt.xlabel("Advertisement budget")
plt.ylabel("Sales impact")
plt.legend()
plt.show()
```

```python
mse=mean_squared_error(y,y_pred)
print(mse)
```

```python
# Create a parabolic data
np.random.seed(42)
x=np.linspace(-5,5,100).reshape(-1,1)
# generate the y values
y=x**3 - 3*(x**2) + 2 + np.random.randn(100,1)*5

# polyomial features
poly = PolynomialFeatures(degree=3)
x_poly = poly.fit_transform(x)

# polyomial features
poly_2 = PolynomialFeatures(degree=2)
x_poly_2 = poly_2.fit_transform(x)

# Run the model
model_1=LinearRegression()
model_1.fit(x_poly, y)

# Run the model
model_2=LinearRegression()
model_2.fit(x_poly_2, y)
# Prediction of model
y_pred = model_1.predict(x_poly)

# Prediction of model
y_pred_2 = model_2.predict(x_poly_2)

plt.scatter(x,y,color='blue',label='actual data')
plt.plot(x, y_pred, color='red',label='polynomial regression line')
plt.plot(x, y_pred_2, color='green',label='Linear regression line')
plt.xlabel("Advertisement budget")
plt.ylabel("Sales impact")
plt.legend()
plt.show()
```

```python
r21=r2_score(y,y_pred)
print(r21)
r22=r2_score(y,y_pred_2)
print(r22)
```

```python
np.random.seed(42)

x = np.linspace(-5, 5, 100).reshape(-1, 1)
y = x**3 - 3*(x**2) + 2 + np.random.randn(100, 1) * 5 

degrees = [1, 2, 3, 4, 5, 6]

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Actual Data')

for degree in degrees:
    poly = PolynomialFeatures(degree=degree)
    x_poly = poly.fit_transform(x)
    
    model = LinearRegression()
    model.fit(x_poly, y)
    
    y_pred = model.predict(x_poly)
    
    plt.plot(x, y_pred, label=f'Poly Degree {degree}')

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.title("Polynomial Regression with Different Degrees")
plt.show()

theta = np.linspace(0, 2 * np.pi, 100)
r = 5
x_circle = r * np.cos(theta) + np.random.randn(100) * 0.3
y_circle = r * np.sin(theta) + np.random.randn(100) * 0.3

degrees = [2, 3, 4, 5, 6]

plt.figure(figsize=(6, 6))
plt.scatter(x_circle, y_circle, color='purple', label="Actual Circular Data")

for degree in degrees:
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(np.column_stack((x_circle, y_circle)))

    model = LinearRegression()
    model.fit(X_poly, y_circle)

    y_pred = model.predict(X_poly)

    plt.scatter(x_circle, y_pred, label=f'Poly Degree {degree}', alpha=0.6)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Polynomial Regression on Circular Data")
plt.legend()
plt.axis("equal")  # Keep aspect ratio equal
plt.show()
```
